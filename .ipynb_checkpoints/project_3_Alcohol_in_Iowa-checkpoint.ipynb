#  Market research for new alcohol store locations

The scenario of this project is to help an alocohol retailer to determine where to open its next store in Iowa, United States. This project mainly looks into predictive modelling using regression techniques. The data used is 10% of the total alcohol sales that took place in Iowa between 2015 and 2016. Each instance is a record of the transaction including information regarding the date(time), location, alcohol type and amount (both in terms of volume and costs). 

#### Problem Statement

The alcohol company would like to open a new store in Iowa to boost its precense and market share in the state. However there are many different locations in Iowa and business environment are very different. The goal of the analysis is to identify potential locations for new storefronts to be set up.

### Fill in counties

From the info above we can see that about 1000 county entries are missing. However using the database from http://www.unitedstateszipcodes.org/ia/ it is possible to fill in the missing values by interpreting the zip code.

```python
zipcode = pd.read_csv('zip_code_database.csv')
zipcode.head()
```

```python
zipcode = zipcode[zipcode['state']=='IA']
zipcode = zipcode[['zip', 'primary_city', 'county']]
zipcode.head()
```

```python
zipcode.set_index('zip', inplace=True)
zip_dic = zipcode.to_dict()
zip_dic = zip_dic['county']
zip_dic
```

```python
zip_dic = zipcode.to_dict()
zip_dic = zip_dic['county']
liquor['Zip Code'] = liquor['Zip Code'].astype(int)
liquor['County'] = liquor['Zip Code'].map(zip_dic)
liquor.loc[liquor['County'].isnull(), 'City'].value_counts()
```

```python
cities = ['Manchester', 'Davenport', 'Cedar Rapids', 'Dunlap', 'Des Moines', 'Burlington']
for c in cities:
    print zipcode[zipcode['primary_city']==c]
```

```python
def fillcounty(x):
    cities = ['Manchester', 'Davenport', 'Cedar Rapids', 'Dunlap', 'Des Moines', 'Burlington']
    counties = ['Delaware County', 'Scott County', 'Linn County', 'Harrison County', 'Polk County', 'Des Moines County']
    for i in range(len(cities)):
        if x == cities[i].upper():
            return counties[i]
        
liquor.loc[liquor['County'].isnull(), 'County'] = liquor.loc[liquor['County'].isnull(), 'City'].apply(fillcounty)
```

### For each county

Using linear regression, we try to find out the relationship between sales sum with other variables. They are average bottle retail price and the number of bottles sold resoectively
#### Price per bottle

```python
reg = LinearRegression()
for i in counties:
    df_criteria = liquor_2015_q1[liquor_2015_q1['County <lambda>'] == i]
    df_target = liquor_2016_q1[liquor_2016_q1['County <lambda>'] == i]
    x_train = df_criteria['State Bottle Retail mean']
    x_train = x_train.reshape(-1,1)
    y_train = df_criteria['Sale (Dollars) sum']
    x_test = df_target['State Bottle Retail mean']
    x_test = x_test.reshape(-1,1)
    y_test = df_target['Sale (Dollars) sum']
    model = reg.fit(x_train,y_train)
    pred = model.predict(x_test)
    print i
    print ''
    print 'r2 score: ', r2_score(y_test, pred)
    print 'mse: ', mean_squared_error(y_test, pred)
    print 'coefficients: ', model.coef_
    print ''
    print '---------------------'
    print''
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.scatter(x_test,y_test)
    ax.plot(x_test,pred)
    ax.set(xlabel='mean bottle price', ylabel='sales sum')
    plt.show()
```

### Conclusions


Drawing a conclusion from the analysis above, we can try to come up with a few suggestions on where it might be a good idea to open our new store front. Ideally the new store front should be located in a place where total sales is high. Depending on the number of stores that we would like to open, there might be different alternatives. If we are only going to open 1 store, it is the best idea to zoom in to the 'Zip Code' scale to continue the analysis. 3 out of the top 5 zip codes in terms of sales per store are located in Des Moines, the capital of Iowa. Therefore it should be a good option if we have not already got a store in the capital.

```python
df_criteria = liquor_2015_q1[liquor_2015_q1['City <lambda>'] == 'DES MOINES']
df_target = liquor_2016_q1[liquor_2016_q1['City <lambda>'] == 'DES MOINES']  
x_train = df_criteria['Volume Sold (Liters) sum']
x_train = x_train.reshape(-1,1)
y_train = df_criteria['Sale (Dollars) sum']
x_test = df_target['Volume Sold (Liters) sum']
x_test = x_test.reshape(-1,1)
y_test = df_target['Sale (Dollars) sum']
scores = cross_val_score(reg, x_train,y_train, cv=8)
model = reg.fit(x_train,y_train)
pred = model.predict(x_test)
print 'Des Moines'
print ''
print 'r2 score: ', r2_score(y_test, pred)
print 'mse: ', mean_squared_error(y_test, pred)
print 'coefficients: ', model.coef_
print 'cv scores: ', scores
print ''
print '---------------------'
print''
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.scatter(x_test,y_test)
ax.plot(x_test,pred)
ax.set(xlabel='Total volume sold', ylabel='sales sum')
plt.show()
```

The above plot shows the total volume sold against total sales. It is very intuitve to figure out that the greater the slope is the better. Another way to increase accuracy of prediction is to include more variable. The following example demonstrate how it is done. However one thing to keep in mind is that some predictors variable are correlated with each other and would hamper the performance of the overall model. A way to limit this problem is to introduce regularised regression such as the Lasso regression shown in the example. It works by limiting the size of the coefficients of variables (someitimes outright reducing them to 0) to prevent the correlated predictors from affecting the model more than they should. How much limiting effect is enforced is determined by the alpha level.

```python
reg = LassoCV(alphas=np.logspace(-4,4,9))
for i in counties:
    df_criteria = liquor_2015_q1[liquor_2015_q1['County <lambda>'] == i]
    df_target = liquor_2016_q1[liquor_2016_q1['County <lambda>'] == i]
    x_train = df_criteria[['State Bottle Retail mean','Volume Sold (Liters) sum','Bottles Sold sum']]
    y_train = df_criteria['Sale (Dollars) sum']
    x_test = df_target[['State Bottle Retail mean','Volume Sold (Liters) sum','Bottles Sold sum']]
    y_test = df_target['Sale (Dollars) sum']
    model = reg.fit(x_train,y_train)
    pred = model.predict(x_test)
    scores = cross_val_score(reg,x_train,y_train,cv=5)
    print i
    print ''
    print 'r2 score: ', r2_score(y_test, pred)
    print 'mse: ', mean_squared_error(y_test, pred)
    print 'cv scores: ',scores
    print 'Coefficients: ',model.coef_
    print ''
    print '---------------------'
    print''
```
